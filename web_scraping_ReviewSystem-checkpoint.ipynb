```python
#import libraries

import pandas as panda
import requests
from bs4 import BeautifulSoup
```

```python
# send request

url = 'https://www.bankbazaar.com/reviews.html'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html.parser')
```

```python
#read review text
rev_txt = []
review_text_ele = soup.find_all(class_ ='text_here review-desc-more')

for item in review_text_ele:
    rev_txt.append(item.text)
    
```

```python
#read auther name
user_name = []
user_name_elem = soup.find_all(class_='js-author-name')
for item in user_name_elem:
    user_name.append(item.text)
```

```python
#read bank name
bank_name=[]
bank_name_ele = soup.find_all('div',{'itemprop':'itemReviewed'})

for item in bank_name_ele:
    bank_name.append(item.find('img').get('alt'))
```

```python
#create dataframe

final_array =[]
for text,user,bank in zip(rev_txt,user_name,bank_name):
    final_array.append({'Review_Text':text,'User':user,'Bank':bank})
    
df=panda.DataFrame(final_array)
df
```
